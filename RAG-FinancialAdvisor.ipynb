```python
# Add Azure OpenAI package
from openai import AzureOpenAI
```

```python
import json

# Flag to show citations
show_citations = True

# Get configuration settings
azure_oai_endpoint =""
azure_oai_key =""
azure_oai_deployment =""
azure_search_endpoint = ""
azure_search_key = ""
azure_search_index = ""
```

```python
# Initialize the Azure OpenAI client
client = AzureOpenAI(
            base_url=f"{azure_oai_endpoint}/openai/deployments/{azure_oai_deployment}/extensions",
            api_key=azure_oai_key,
            api_version="2023-09-01-preview")
```

```python
# Configure your data source
extension_config = dict(dataSources = [
        {
            "type": "AzureCognitiveSearch",
            "parameters": {
                "endpoint":azure_search_endpoint,
                "key": azure_search_key,
                "indexName": azure_search_index,
            }
        }]
    )
```

```python
# Get the prompt

#1- Give me financial information related to TKO GROUP HOLDINGS, INC
#2-show me the financial trend over the time for TKO GROUP HOLDINGS, INC
#3-show me the financial trend over the time for TKO GROUP HOLDINGS, INC in the tabular format

#4-compare the financials of RayzeBio,Inc vs TKO GROUP HOLDINGS, INC in tabular format for the available financials figure
#5-compare the financials of HOWARD HUGHES HOLDINGS INC vs TKO GROUP HOLDINGS, INC

#6-based on Account Receivable where do you advice me to do my investment is in MAPLEBEAR INC or TKO GROUP HOLDINGS, INC
text = input('\nEnter a question:\n')
```

```python
# Send request to Azure OpenAI model
print("...Sending the following request to Azure OpenAI endpoint...")
print("Request: " + text + "\n")

response = client.chat.completions.create(
    model = azure_oai_deployment,
    temperature = 0.5,
    max_tokens = 1000,
    messages = [
         {"role": "system", "content": "You are an Financial advisor that helps people find information."},
         {"role": "user", "content": text}
            ],
            extra_body = extension_config
        )
```

```python
# Print response
print("Response: " + response.choices[0].message.content + "\n")
```

```python
#Check for the citations
if (show_citations):
    # Print citations
    print("Citations:")
    citations = response.choices[0].message.context["messages"][0]["content"]
    citation_json = json.loads(citations)
    for c in citation_json["citations"]:
        print("  Title: " + c['title'] + "\n    URL: " + c['url'])
```
